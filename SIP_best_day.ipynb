```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Convert to DataFrame
df = pd.DataFrame(march)

# Convert 'Date' column to datetime format
df['Date '] = pd.to_datetime(df['Date '])

# Set Seaborn style
sns.set_style("whitegrid")

# Create line plot
plt.figure(figsize=(10, 5))
sns.lineplot(x=df['Date '], y=df['Close '], marker='o', color='b', label='Close Price')

# Formatting the graph
plt.xlabel('Date', fontsize=12)
plt.ylabel('Close Price', fontsize=12)
plt.title('Stock Market Trend in Month', fontsize=14)
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def plot_stock_trend(df, month, column='Close '):
    """
    Plots the stock market trend for a given month.

    Parameters:
    df (DataFrame): The stock market dataset with a 'Date' column.
    month (int): The month to filter (1 for January, 2 for February, etc.).
    column (str): The column to plot (default is 'Close').

    Returns:
    None (Displays a Seaborn line plot)
    """

    # Filter data for the given month
    df_filtered = df[df['month'] == month].copy()

    # Ensure 'Date' is in datetime format
    df_filtered['Date '] = pd.to_datetime(df_filtered['Date '])

    # Set Seaborn style
    sns.set_style("whitegrid")

    # Create line plot
    plt.figure(figsize=(10, 5))
    sns.lineplot(x=df_filtered['Date '], y=df_filtered[column], marker='o', color='b', label=f'{column} Price')

    # Formatting the graph
    plt.xlabel('Date', fontsize=12)
    plt.ylabel(f'{column} Price', fontsize=12)
    plt.title(f'Stock Market Trend in Month {month}', fontsize=14)
    plt.xticks(rotation=45)
    plt.legend()
    plt.show()
```

```python
# Call function for March (month = 3)
plot_stock_trend(nifty, month=2)
```
